### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
```

```python
dist = sp.stats.gamma(a=1, loc=0, scale=1)
sample = dist.rvs(size=10000)
plt.figure(figsize=(8,4))
plt.xlim(0,10)
plt.xticks(np.arange(0,10.5,1))
plt.hist(sample, bins=50, density=True, label='distribuzione gamma')
x = np.linspace(0, 10)
plt.plot(x, dist.pdf(x), label='PDF')
plt.plot(x, dist.cdf(x), label='CDF')
plt.legend()
plt.show() 
```

```python
a, loc, scale = sp.stats.gamma.fit(sample)
a, loc, scale
```

### **2. Fitta i dati**
Prova a fittare i dati sottostanti con le migliori curve, calcola il MAE e l'RMSE

```python
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
months = np.arange(12)
plt.figure(figsize=(12,6))
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")

x_interp = np.linspace(0, 11, 100)

# Interpolazione lineare Scipy
f_lin_min = sp.interpolate.interp1d(months, temp_min, kind='linear')
y_lin_min = f_lin_min(x_interp)

f_lin_max = sp.interpolate.interp1d(months, temp_max, kind='linear')
y_lin_max = f_lin_max(x_interp)

plt.plot(x_interp, y_lin_min, color='green', label='Interpolazione lineare Scipy')
plt.plot(x_interp, y_lin_max, color='darkorange', label='Interpolazione lineare Scipy')


# Interpolazione spline cubica Scipy
f_cubic_min = sp.interpolate.interp1d(months, temp_min, kind='cubic')
y_cubic_min = f_cubic_min(x_interp)

f_cubic_max = sp.interpolate.interp1d(months, temp_max, kind='cubic')
y_cubic_max = f_cubic_max(x_interp)

plt.plot(x_interp, y_cubic_min, linestyle='-.', color='red', label='Interpolazione cubica Scipy')
plt.plot(x_interp, y_cubic_max, linestyle='--',color='blue', label='Interpolazione cubica Scipy')


mae_min = np.mean(np.abs(temp_min - f_cubic_min(months)))
rmse_min = np.sqrt(np.mean((temp_min - f_cubic_min(months))**2))
mae_max = np.mean(np.abs(temp_max - f_cubic_max(months)))
rmse_max = np.sqrt(np.mean((temp_max - f_cubic_max(months))**2))

print(f"MAE min: {mae_min} \n RMSE min: {rmse_min} \n MAE max: {rmse_max} \n RMSE max: {rmse_max}")

plt.legend()
plt.show()
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = df["mpg"].values
y = df["disp"].values

# Regressione lineare con SciPy
slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x, y)

print(f"R quadro = {r_value**2}")
print(f"P-value = {p_value}")
print(f"Standard error = {std_err}")

y_pred = intercept + slope * x

# Plot
plt.scatter(x, y, color='blue', label='Dati')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regressione')
plt.xlabel('Miles per gallon (mpg)')
plt.ylabel('Displacement')
plt.legend()
plt.show()
```
